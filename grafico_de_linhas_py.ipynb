Visualização de dados em Python

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 5]
y = [2, 3, 7]
```

```python
#LegendasTítulo
plt.title("Meu primeiro gráfico em Python")

#Legendas eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]

titulo = "Gráfico de barras"
eixox = "Eixo X"
eixoy = "Eixo Y"

#Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x, y)
```
